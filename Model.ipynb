```python
import numpy as np
import pandas as pd
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
```

```python
def import_from_csv(dir,actions=[]):
    filenames = [f for f in listdir(dir)
        if isfile(join(dir, f))]
    res={}
    for f in filenames:
        a="_".join(f.split("_")[:-1])
        if (a in actions) or not actions:
            df=pd.read_csv(join(dir, f))
            if a not in res.keys():
                res[a]=[]
            res[a].append(df)
    return res
```

```python
actions=["walk","nothing"]
cwt_train=import_from_csv('cwt_train/',actions)

for a in cwt_train.keys():
    print(a, len(cwt_train[a]))
cwt_train["walk"][-1]
```

```python
cwt_test=import_from_csv('cwt_test/',actions)

for a in cwt_test.keys():
    print(a, len(cwt_test[a]))
cwt_test["walk"][-1]
```

```python
plt.figure(figsize=(30,4))
counts_for_lengths={}

for a in actions:
    plot_x=[]
    plot_y=[]
    for df in cwt_train[a]:
        x=df.shape[1]
        if x not in counts_for_lengths.keys():
            counts_for_lengths[x]=1
            y=1
        else:
            y=counts_for_lengths[x]+1
            counts_for_lengths[x]+=1
        plot_x.append(x)
        plot_y.append(y)
    print(a)
    plt.scatter(plot_x,plot_y,label=a,s=8)

print(sum(counts_for_lengths.values()))
plt.legend()
plt.show()
```

Adding cuts from "nothing" to get same shapes

```python
def to_same_sizes(data,l,r,filler):
    res={}
    for a in data.keys():
        res[a]=[]
        for df in data[a]:
            if df.shape[1]==r:
                #print(1, df.shape)
                res[a].append(df)
                continue
            if df.shape[1]>r:
                df_=df.iloc[:,:r]
                #print(2, df_.shape)
                res[a].append(df_)
                continue
            if df.shape[1]<r and df.shape[1]>l:
                df_=pd.concat( [ df.iloc[:,:r], filler.iloc[:,:(r-df.shape[1])] ] ,axis=1)
                #print(3, df_.shape)
                res[a].append(df_)

    return res
```

```python
l=300
r=357
nothing_df=cwt_train["nothing"][-1]
sh1,sh2=nothing_df.shape
print(sh1,sh2)
cwt_train["nothing"]=cwt_train["nothing"][:-1]

trainDct=to_same_sizes(cwt_train,l,r,nothing_df)
testDct=to_same_sizes(cwt_test,l,r,nothing_df)

for a in actions:
    print(a, len(trainDct[a]),len(testDct[a]))
```

```python
def to_one_df(data,action_dct):
    res=pd.DataFrame()
    for action in data.keys():
        for df in data[action]:
            row=(df.values).flatten()
            row=np.append(row,action_dct[action])
            res=res.append(pd.Series(row),ignore_index=True)
            print(res.shape)
    return res
```

```python
action_dct={}
i=0
for a in trainDct.keys():
    action_dct[a]=i
    i+=1

train=to_one_df(trainDct,action_dct)
```

```python
test=to_one_df(testDct,action_dct)
test
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
test=test.rename(columns={test.columns[-1]: "target"})
train=train.rename(columns={train.columns[-1]: "target"})
```

```python
train_df=train.drop(columns=["target"])
train_target=train["target"]
test_df=test.drop(columns=["target"])
test_target=test["target"]
```

```python
rf = RandomForestClassifier(oob_score=True, random_state=1, n_jobs=-1)
```

```python
param_grid = { "criterion" : ["entropy"],
               "min_samples_leaf" : [1, 5, 10],
               "min_samples_split" : [2, 4, 10],
               "max_depth": [6, 8, 10],
               "n_estimators": [50, 100, 200]}
gs = GridSearchCV(estimator=rf, param_grid=param_grid, scoring='accuracy',
                  cv=3, verbose=10)
gs = gs.fit(train_df, train_target)
```

```python
print(gs.best_score_)
print(gs.best_params_)

best_est = gs.best_estimator_

print(best_est.score(test_df, test_target))
```
